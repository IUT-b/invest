```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
start='2019-06-01'
end='2020-06-01'

df=data.DataReader('^N225','yahoo',start,end)
```

```python
date=df.index
price=df['Adj Close']
```

```python
plt.plot(date,price)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.title('N225',color='blue',backgroundcolor='white',size=40,loc='center')
plt.xlabel('date',color='black',size=30)
plt.ylabel('price',color='black',size=30)
plt.legend()
```

```python
span01=5
span02=25
span03=50

df['sam01']=price.rolling(window=span01).mean()
df['sam02']=price.rolling(window=span02).mean()
df['sam03']=price.rolling(window=span03).mean()
```

```python
pd.set_option('display.max_rows',None)
df.head(100)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.plot(date,df['sam01'],label='sam01')
plt.plot(date,df['sam02'],label='sam02')
plt.plot(date,df['sam03'],label='sam03')
plt.title('N225',color='blue',backgroundcolor='white',size=40,loc='center')
plt.xlabel('date',color='black',size=30)
plt.ylabel('price',color='black',size=30)
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.subplot(2,1,1)
plt.plot(date,price,label='Close')
plt.plot(date,df['sam01'],label='sam01')
plt.plot(date,df['sam02'],label='sam02')
plt.plot(date,df['sam03'],label='sam03')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend()
```
